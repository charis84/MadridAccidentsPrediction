```
from rdflib import Graph
```

```
g = Graph()
result = g.parse("MadridAccidentsPrediction/Accidentalidad_Madrid_2018_FINAL.ttl", format="n3")

print("Graph has %s statements" % len(g))
```

```
qres = g.query(
    """SELECT DISTINCT ?evntType
        WHERE {
          ?evntType acdt:accidentType "ATROPELLO"
        }
""")

for row in qres:
  print(row.labels)
```
